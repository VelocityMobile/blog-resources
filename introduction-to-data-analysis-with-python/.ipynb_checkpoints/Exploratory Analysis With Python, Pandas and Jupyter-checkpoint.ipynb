# Exploratory Analysis With Python, Pandas and Jupyter

At Velocity, being able to explore, visualize, and analyze our data is paramount to our business. We use a variety of tools to gain insight from our data, such as [Grow](https://www.grow.com/), Google Sheets, SQL queries, and also a few other custom in-house tools.

One particular set of tools, Python, Pandas, and Jupyter, I think, will play a very important role in our data analysis toolkit.
 
Since I can’t use our customer data to illustrate, I’ll use a dataset scraped from the city archive site of my home city of Edmonton in Canada!

This dataset contains a list of photos taken throughout Edmonton at various points in time. Each row contains some metadata about the photo, such as title, approximate date, subject tags, etc…

In this post, we will explore a couple of aspects of this dataset; namely the distribution of photos by year and the frequency of subject tags associated with the photos.

Inside our Jupyter notebook, we will first load the libraries we will be using:

````python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

from IPython.display import Image, display
from IPython.core.display import HTML 
````

Next step will be to load the CSV dataset using Pandas. It can be done like so:

````python
photos = pd.read_csv('data/yeg_city_archive.csv')
photos.head()
````

Since we want to look at distribution of photos by year, it looks like we will need to clean up the ```creation_date``` column! If we look at all the different values of the ```creation_date``` column, it appears the ```creation_date``` will be in one of the following formats:

1. year  
2. [ca. year]  
3. c. year  
4. year-month-day  
5. Month day, year

The year is always in a 4-digit format, so we can use a simple regular expression to extract the year.


````python
import re
year_pattern = re.compile('[0-9]{4}')

def parse_year(date):
    if type(date) is not str:
        return None
    
    match = year_pattern.search(date)
    
    return int(match.group()) if match is not None else None
````

Next, we will apply the regular expression to the ```creation_date``` column:

````python
photos['creation_date_year'] = photos['creation_date'].apply(parse_year)
photos['creation_date_year'].head()
````

Beautiful! We now have the year as a number.

Now that we have the data with the year extracted, we can begin to analyze the distribution of photos by year. One way of doing this is to group the photos by decade. Let's first start by taking a look at the minimum year and the maximum year:

````python
min_year = photos['creation_date_year'].min()
max_year = photos['creation_date_year'].max()

print('Minimum Year: {0}\nMaximum Year: {1}'.format(min_year, max_year))
````

If we wanted to break this year range into decades, let's see how many categories would be required:

````python
lower_bound = min_year - (min_year % 10)
upper_bound = max_year - (max_year % 10) + 10 if max_year % 10 > 0 else max_year
number_of_categories = (upper_bound - lower_bound) / 10

print('Number of Categories required: {0}'.format(number_of_categories))
````

14 categories can easily be displayed on a chart, so we will continue down this route and categorize each photo by year:

````python
year_categories = list(range(int(lower_bound), int(upper_bound) + 10, 10))
photos['creation_date_cat'] = pd.cut(photos.creation_date_year, year_categories)
````

Now let's create a bar chart to display the distribution of photos by year:

````python
groups = photos.groupby('creation_date_cat').size()
p = groups.plot(kind='bar', figsize=(11, 7))
p.set(xlabel="Creation Year", ylabel="Number of Photos", title="Photos By Year")
rects = p.patches
labels = groups.values

for rect, label in zip(rects, labels):
    height = rect.get_height()
    p.text(rect.get_x() + rect.get_width()/2, height + 5, label, ha='center', va='bottom')
````

Based on this chart, it appears the vast majority of the photos are within the 2010 - 2020 date range, which isn't the most interesting. However, there are still a fair amount of photos that are pre-21st century. Let's take a look at some of the photos from the 19th century:

````python
oldest_photos = photos[photos['creation_date_year'] <= 1890]['image_url'].values

for photo_url in oldest_photos:
    display(Image(url = photo_url)) # Download image and display in notebook
````

Another point of interest is analyzing the subject tags associated with each of the photos. A fair amount of the photos have one or more subject tags. It would be neat to visualize the subject tags in a word cloud.  
  
We will need to create a dictionary that will contain the tag as the key and its associated count as the value. We can then pass this dictionary to the word cloud function. 
    
The subject column contains a list of the associated subject tags. The tags associated with each photo will be in a comma-delimited string, so we will need to split the string. This operation will result in a Series containing a list of subject tags. We cannot do too much with this Series in this state, so we will need to flatten the Series. After that, we can get a count of each of the subject tags.

Finally, we will only look at subject tags that occur at least 5 times.

````python
photos['subjects'] = photos['subjects'].astype(np.str)

# Split tags by comma, and strip leading/trailing whitespace
subjects_list = photos['subjects'].apply(lambda x: list(map(lambda subject: subject.strip(), x.split(',')))).values
# Flatten the list of tag lists
subjects_s = np.hstack(subjects_list.flat)
# Construct a data frame
subjects_df = pd.DataFrame({'subjects': subjects_s})
# Count the number of each of the tags
subjects_grp = subjects_df.groupby('subjects').size().sort_values(ascending=False)

# Ignore photos with no subject
subjects_grp_filtered = subjects_grp[2:]

# Only include subjects that are assigned to at least 5 photos
subjects = subjects_grp_filtered[subjects_grp_filtered > 5]
````

Now that we have the subject tag counts, we can create the frequency table / dictionary. We can pass this dictionary to the word cloud object and generate the actual word cloud image.

````python
frequencies = dict(zip(list(subjects.index), list(subjects.values)))
wordcloud = WordCloud(max_font_size=40, relative_scaling=0.5, scale=2.5).generate_from_frequencies(frequencies)

plt.figure(figsize=(16, 6))
plt.imshow(wordcloud)
plt.axis("off")
plt.show();
````

Based on this result, it appears that a fair amount of the photos are aerial views of the city, followed by cityscapes, trees, and clouds. I am kind of surprised by the frequency of the 'clouds' tag though. I would have thought historic buildings would have had a higher frequency.

As you can see, using Python + Pandas allows you to easily clean, slice n’ dice, and analyse your data in a very straightforward fashion. When you add a Jupyter notebook into the mix, you can embed images, charts, and many other visualizations.
